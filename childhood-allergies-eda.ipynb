### About this dataset

**This dataset contains the power to help us better understand the prevalence and treatment outcomes of childhood allergies over an extended period of time. Not only does it publicize the number of individuals currently suffering from asthma, atopic dermatitis, allergic rhinitis and food allergies through retrospective data as reported by healthcare providers - but it also features a set of columns which allow us to gain valuable insights into how these outcomes differ across different demographics such as gender, race and ethnicity. By further examining this data, we can start to recognize patterns in trends among the diagnosed cases - paving way for new treatments and prevention strategies that could prevent severe allergic reactions for many children all around the world**

```python
df = pd.read_csv(r'D:/food-allergy-analysis-Zenodo.csv')
```

```python
df.head().T
```

```python
# Clearing the dataset

df['GENDER_FACTOR'] = df['GENDER_FACTOR'].str[5:]
df['RACE_FACTOR'] = df['RACE_FACTOR'].str[5:]
df['ETHNICITY_FACTOR'] = df['ETHNICITY_FACTOR'].str[5:]
df['PAYER_FACTOR'] = df['PAYER_FACTOR'].str[5:]
df
```

```python
df.info()
```

```python
# Excluding negative values

negative_values = df[df.ATOPIC_DERM_START < 0].index

df.drop(negative_values, inplace=True)
```

```python
# Viewing the allergies


# function to add value labels
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center', va = 'bottom', c = 'blue')


plt.figure(figsize=(16, 7))
x=['SHELL FISH', 'FISH ALG', 'MILK_ALG', 'SOY_ALG', 'EGG_ALG', 'WHEAT_ALG', 'PEANUT_ALG',
   'SESAME_ALG', 'TREENUT_ALG', 'WALNUT_ALG', 'PECAN_ALG', 'PISTACH_ALG', 'ALMOND_ALG',
   'BRAZIL_ALG', 'HAZELNUT_ALG', 'CASHEW_ALG', 'ATOPIC_DERM', 'ALLERGIC_RHINITIS', 'ASTHMA']
height=[df["SHELLFISH_ALG_START"].count(),df["FISH_ALG_START"].count(),
       df["MILK_ALG_START"].count(),df["SOY_ALG_START"].count(),df["EGG_ALG_START"].count(),
       df["WHEAT_ALG_START"].count(),df["PEANUT_ALG_START"].count(),df["SESAME_ALG_START"].count(),
       df["TREENUT_ALG_START"].count(),df["WALNUT_ALG_START"].count(),df["PECAN_ALG_START"].count(),
       df["PISTACH_ALG_START"].count(),df["ALMOND_ALG_START"].count(),df["BRAZIL_ALG_START"].count(),
       df["HAZELNUT_ALG_START"].count(),df["CASHEW_ALG_START"].count(),df["ATOPIC_DERM_START"].count(),
       df["ALLERGIC_RHINITIS_START"].count(),df["ASTHMA_START"].count()]
plt.bar(x=x, height=height)

addlabels(x, height)
plt.xticks(rotation=60)
plt.title('Number of allergy cases', fontsize=18)
plt.box(True)
plt.show()
```

```python
# Create a dataframe (df_ad) with only ATOPIC_DERM records

df_ad = df[['BIRTH_YEAR', 'GENDER_FACTOR', 'RACE_FACTOR', 'ETHNICITY_FACTOR', 'PAYER_FACTOR',
           'ATOPIC_MARCH_COHORT', 'AGE_START_YEARS', 'AGE_END_YEARS', 'ATOPIC_DERM_START', 'ATOPIC_DERM_END']]
df_ad = df_ad.dropna(subset=["ATOPIC_DERM_START"])

df_ad
```
